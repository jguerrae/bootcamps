## SIGNIFICADO INTUITIVO DE LAS COLUMNAS
<ol>
<li> <strong> v_estension </strong>: no entiendo </li> <br>

<li> <strong> location </strong>: *Diccionario* con el nombre del país, estado y ciudad. Parece ser uniforme a toda la muestra. </li> <br>

<li> <strong> job_title </strong>: *String* con el nombre del titulo. Parece ser uniforme. Este es el titulo del trabajo que tienen actualmente. De aquí se podría extraer el tipo de empleo. ¿Cuáles serían esos tipos de empleos? </li> <br>

<li> <strong> company </strong>: *Diccionario* que contiene información sobre la compañia en la cuál trabaja. Parece ser uniforme. El campo pertenece a perfil de linkedin de la compañia. </li> <br>

<li> <strong> job_history </strong>: Lista con diccionarios que contiene el historial de empleo. El primer diccionario parece corresponder al empleo que tienen en la actualidad. Si no tiene "end" es porque probablemente así sea. </li> <br>

<li> <strong> snapshots </strong>: *Diccionarios* contiene el historial de resumenes de la personas en cada momento (año-mes) del tiempo. Es un diccionario que tiene como llave el tiempo y dentro tiene el encabezado del perfil (titulo) la compañia, location. Sin embargo parece no guardar una uniformidad. No sería buena fuente para extraer el historial. </li> <br>

<li> <strong> skills </strong>: *Lista* con resumen de la habilidades. Podría ser útil para saber si la personas es programadora o no. </li> <br>

<li> <strong> education </strong>: Lista de diccionarios con información sobre la educación. Dentro diccionario hay un resumen, nombre de la institución, grado que recibe (aquí tiene dentro una lista -  pero en realidad un string), el año en el que empezó y el año en que terminó. Si el patrón es unirome es posible extraer aquí el potencial bootcamp y su tiempo de terminación. Dado que esto es un string vale la pena eliminar \r y \n. Parece ser uniforme. </li> <br>

<li> <strong> about </strong>: Resumen de la personas. Hay información de todo tipo. Valdría la pena revisar si tienen palabras relacionadas con ciencias de computación. ALGUNOS PONEN ACÁ LA CERTIFICACIÓN DEL BOOTCAMP. REVISAR ALGUNOS CASOS ACÁ  </li> <br>

<li> <strong> updated </strong>: fecha de actualización del perfil </li> <br>

<li> <strong> num_connnections </strong>: *Entero* con numero de contactos. Puede ser quizás una buena proxy de posibilidades de conseguir empleo. </li> <br>

<li> <strong> interests </strong>: *Lista *con los intereses. También puede ser una buena fuente para determinar si la persona está dentro de la CC. </li> <br>

<li> <strong> volunteered </strong>: información y extructura heterogenea. </li> <br>

<li> <strong> certifications </strong>: Misma estructura que educación. Puede ser muy útil. Hasta el momento educación y certificación son las dos columnas para buscar el bootcamp en el cual estuvo la persona </li> <br>

<li> <strong> industry </strong>: *string* con la industria, puede ser útil pero parece tener muchos NAs </li> <br>
